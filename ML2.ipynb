```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import shapiro

#  Load from external CSV (UCI format)
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv("iris.csv", header=None, names=col_names)
```

1.	Create a histogram to visualize the distribution of sepal length across all Iris samples.

```python
plt.figure(figsize=(16,6))
plt.hist(iris['sepal_length'], bins=15, color='skyblue', edgecolor='black')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")

plt.show()
```

2.	Create histograms of sepal width for each Iris species in a single plot, using different colors for each species.


```python
plt.figure(figsize=(16,5))
species_list = iris['species'].unique()
colors = ['red', 'green', 'blue']
for sp, color in zip(species_list, colors):
    subset = iris[iris['species'] == sp]
    plt.hist(subset['sepal_width'], bins=10, alpha=0.5, label=sp, color=color, edgecolor='black')
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Width by Species")
plt.legend()
plt.show()
```

3.	Create two histograms in a single figure (using subplots) to compare the distributions of petal length and petal width.


```python
fig, axes = plt.subplots(1, 2, figsize=(25,6))
axes[0].hist(iris['petal_length'], bins=10, color='orange', edgecolor='black')
axes[0].set_title("Petal Length Distribution")
axes[0].set_xlabel("Petal Length (cm)")
axes[0].set_ylabel("Frequency")

axes[1].hist(iris['petal_width'], bins=10, color='purple', edgecolor='black')
axes[1].set_title("Petal Width Distribution")
axes[1].set_xlabel("Petal Width (cm)")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(24, 8))

# Histogram for petal length
sns.histplot(iris['petal_length'], bins=20, kde=True, color='orchid', edgecolor='black', ax=axes[0])
axes[0].set_title('Distribution of Petal Length')
axes[0].set_xlabel('Petal Length (cm)')
axes[0].set_ylabel('Frequency')

# Histogram for petal width
sns.histplot(iris['petal_width'], bins=20, kde=True, color='teal', edgecolor='black', ax=axes[1])
axes[1].set_title('Distribution of Petal Width')
axes[1].set_xlabel('Petal Width (cm)')
axes[1].set_ylabel('Frequency')

# Final layout
plt.tight_layout()
plt.show()
```

4.	Create a boxplot to compare sepal length across the three Iris species.


```python
plt.figure(figsize=(8,10))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Sepal Length by Species")
plt.show()
```

5.	Create a single figure with four boxplots comparing sepal length, sepal width, petal length, and petal width across all samples (not by species).


```python
fig, axes = plt.subplots(2, 2, figsize=(10,14))
sns.boxplot(y=iris['sepal_length'], ax=axes[0,0])
axes[0,0].set_title("Sepal Length")
sns.boxplot(y=iris['sepal_width'], ax=axes[0,1])
axes[0,1].set_title("Sepal Width")
sns.boxplot(y=iris['petal_length'], ax=axes[1,0])
axes[1,0].set_title("Petal Length")
sns.boxplot(y=iris['petal_width'], ax=axes[1,1])
axes[1,1].set_title("Petal Width")
plt.tight_layout()
plt.show()
```

6.	Create a boxplot of petal width, grouped by species, and use hue to further differentiate by species (or another feature if applicable).


```python
plt.figure(figsize=(6,5))
sns.boxplot(x='species', y='petal_width', hue='species', data=iris)
plt.title("Petal Width by Species")
plt.show()
```

7.	Create a scatterplot to compare petal length and petal width, colored by species.


```python
plt.figure(figsize=(15,5))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris, palette='deep')
plt.title("Petal Length vs Petal Width")
plt.show()
```

8.	Create a scatterplot matrix (pair plot) to visualize relationships between all four measurements (sepal length, sepal width, petal length, petal width) with species differentiation.


```python
import seaborn as sns
import pandas as pd

# Load CSV
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv("iris.csv", header=None, names=col_names)

# Remove rows with missing values
iris = iris.dropna()

# Convert first 4 columns to numeric (in case they are strings with spaces)
for col in col_names[:4]:
    iris[col] = pd.to_numeric(iris[col], errors='coerce')

# Double-check data types
print(iris.dtypes)

# Now create pair plot
sns.pairplot(iris, hue='species', diag_kind='kde', palette='Set2')
```
