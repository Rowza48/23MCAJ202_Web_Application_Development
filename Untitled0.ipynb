6.Find a matrix or vector norm using Numpy

```python
vector = np.array([1, 2, 3, 4, 5])
vector_norm = np.linalg.norm(vector)
print("Vector:", vector)
print("Vector Norm (L2):", vector_norm)
matrix = np.array([[1, 2],
                   [3, 4]])
matrix_norm = np.linalg.norm(matrix)
print("\nMatrix:")
print(matrix)
print("Matrix Norm (Frobenius):", matrix_norm)
vector_l1_norm = np.linalg.norm(vector, ord=1)
print("\nVector L1 Norm:", vector_l1_norm)
matrix_inf_norm = np.linalg.norm(matrix, ord=np.inf)
print("Matrix L-infinity Norm:", matrix_inf_norm)
```

3.Plot a horizontal line in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.axhline(y=0.5, color='r', linestyle='-', label='Horizontal Line at y=0.5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot with a Horizontal Line')
plt.legend()
plt.grid(True)
plt.show()
```

6.How to display the value of each bar in a bar chart using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
fig,ax = plt.subplots(figsize=(8,5))
bars= ax.bar(categories, values)
ax.bar_label(bars)
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval,2),
    ha='center', va='bottom')
    ax.set_title('Bar Chart with Values')
    plt.show()
```

7. How to annotate bars in barplot with matplotlib in python

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
fig,ax=plt.subplots()
bars=ax.bar(categories, values,color="skyblue")
for bar in bars:
  height=bar.get_height()
  ax.annotate(f'{height}',
  xy=(bar.get_x() + bar.get_width() / 2, height),
  xytext=(0, 3),
  textcoords="offset points",
  ha='center', va='bottom')
ax.set_title('Bar Chart with Values')
plt.show()

```

8.How to change the color of a graph plot in matplotlib with python

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
y2=np.cos(x)
plt.plot(x,y,color="red")
plt.plot(x,y2,color="green")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph")
plt.show()
```
